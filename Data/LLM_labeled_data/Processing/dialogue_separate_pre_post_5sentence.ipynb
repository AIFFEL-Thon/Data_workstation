# 2. 데이터 준비

```python
# 데이터 로드
# 발화 분리, 감정이 태깅된 자유대화(성인) inside
file_path = '/content/drive/MyDrive/AIFFELthon/Data/Sample/Sample_감정이태깅된자유대화(성인)/utterance_json_output_inside.csv'  # 업로드된 파일 경로
sample_data = pd.read_csv(file_path)

# 필요한 열만 추출
data = sample_data[['FileName', 'Text']].copy()
```

```python
# 파일명 기준으로 대화 분리
def group_dialogues_by_filename(data):
    # 파일명 기준 그룹화
    data['DialogueGroup'] = data['FileName'].str.extract(r'(.+?)_\d{6}_\d{6}.wav')
    grouped_data = data.groupby('DialogueGroup')['Text'].apply(list).reset_index()
    return grouped_data
```

```python
# 대화 데이터 분리
grouped_data = group_dialogues_by_filename(data)
grouped_data
```

```python
# 필요한 열만 추출
data = sample_data[['FileName', 'Text', 'SpeakerEmotionTarget', 'VerifyEmotionTarget']].copy()

# 파일명 기준으로 대화 분리
def group_dialogues_by_filename_sentences(data):
    # 파일명 기준 그룹화
    data['DialogueGroup'] = data['FileName'].str.extract(r'(.+?)_\d{6}_\d{6}.wav')
    sentences_grouped = data.groupby('FileName').apply(
        lambda group: pd.DataFrame({
            'GroupID': [group.iloc[0]['DialogueGroup']]*len(group),
            'FileName': group['FileName'],
            'Sentence': group['Text'],
            'SpeakerEmotionTarget': group['SpeakerEmotionTarget'],
            'VerifyEmotionTarget': group['VerifyEmotionTarget']
        })
    ).reset_index(drop=True)
    return sentences_grouped
```

```python
# 대화 데이터 분리
sentences_grouped_data = group_dialogues_by_filename_sentences(data)

sentences_grouped_data
```

```python
# GroupID가 같은 것들의 개수를 세기
group_counts = sentences_grouped_data.groupby('GroupID').size()

print(group_counts)
```

```python
grouped_sentences = sentences_grouped_data.copy()
```

# 3. 앞뒤 5문장 추가한 새로운 컬럼 추가

```python
# 앞뒤 5개 문장을 포함한 리스트 생성 함수
def get_pre_post_sentences(group):
    sentences = group['Sentence'].tolist()
    pre_post_sentences = []

    for i in range(len(sentences)):
        # 앞뒤 5개 문장 선택
        start_idx = max(0, i - 5)
        end_idx = min(len(sentences), i + 6)

        # 해당 문장과 앞뒤 문장을 리스트로 생성
        context = sentences[start_idx:end_idx]
        pre_post_sentences.append(context)

    # 원본 데이터프레임에 새로운 리스트 열 추가
    group['pre_post_5_sentences'] = pre_post_sentences
    return group
```

```python
# GroupID로 그룹화하여 처리
result = grouped_sentences.groupby('GroupID').apply(get_pre_post_sentences)
```
